## 1. Introduction to Databases

A **database** is an organized collection of structured data that can be easily accessed, managed, and updated.

Examples include:
- Student Information Systems
- Banking Systems
- E-commerce Platforms
- Healthcare Records

Most modern applications rely on **Relational Database Management Systems (RDBMS)** such as MySQL.

## 2. What is SQL?

SQL (Structured Query Language) is used to communicate with relational databases.

Typical SQL tasks:
- Creating databases and tables
- Inserting records
- Retrieving data
- Updating records
- Deleting records

## 3. Install Python Connector for MySQL

To connect Python with MySQL Server we install **mysql-connector-python**.

```python
# Install connector (run once)
%pip install mysql-connector-python
```

## 4. Import Library

```python
import mysql.connector
```

## 5. Connect to MySQL Server

Connection parameters:
- host = localhost
- user = root
- password = your MySQL password

```python
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="yourpassword"
)

print("Connection successful")
```

## 6. Create Database

We create a database named **training_db**.

```python
cursor = connection.cursor()

cursor.execute("CREATE DATABASE IF NOT EXISTS training_db")

print("Database created")
```

## 7. Connect to training_db

```python
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="yourpassword",
    database="training_db"
)

cursor = connection.cursor()
print("Connected to training_db")
```

## 8. Create students Table

```python
create_table = '''
CREATE TABLE IF NOT EXISTS students (
    student_id INT PRIMARY KEY,
    name VARCHAR(50),
    course VARCHAR(50),
    marks INT
)'''

cursor.execute(create_table)

print("Table created")
```

## 9. Insert Sample Records

```python
insert_query = '''
INSERT INTO students VALUES
(1,'Amit','MCA',85),
(2,'Sara','MCA',90),
(3,'John','MBA',78)
'''

cursor.execute(insert_query)
connection.commit()

print("Data inserted")
```

## 10. Retrieve Data

```python
cursor.execute("SELECT * FROM students")

rows = cursor.fetchall()

for r in rows:
    print(r)
```

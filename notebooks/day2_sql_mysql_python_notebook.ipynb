## 1. Introduction

SQL data manipulation means working with table data using commands such as INSERT, UPDATE, DELETE, and SELECT.

In this Day 2 notebook, the focus is on:
- Inserting new records
- Retrieving complete result sets
- Filtering records based on conditions

## 2. Python Connection

```python
import mysql.connector
```

## 3. Connect Database

Connection parameters:
- host = localhost
- user = root
- password = your MySQL password
- database = training_db

```python
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="yourpassword",
    database="training_db"
)

cursor = conn.cursor()
print("Connected to training_db")
```

## 4. Insert Records

```python
cursor.execute("""
INSERT INTO students VALUES
(4, 'Riya', 'MBA', 88)
""")

conn.commit()
print("Record inserted")
```

## 5. Retrieve Records

```python
cursor.execute("SELECT * FROM students")

for row in cursor.fetchall():
    print(row)
```

## 6. Filtering

```python
cursor.execute("""
SELECT name,marks
FROM students
WHERE marks > 80
""")

for row in cursor.fetchall():
    print(row)
```

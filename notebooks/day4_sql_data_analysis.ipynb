## 1. Connect Database

```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="yourpassword",
    database="training_db"
)

cursor = conn.cursor()
print("Connected to training_db")
```

## 2. Aggregate Analysis

```python
cursor.execute("""
SELECT AVG(amount) FROM sales
""")

result = cursor.fetchone()
print(f"Average Sale Amount: {result[0]}")
```

## 3. Region-wise Group Analysis

```python
cursor.execute("""
SELECT region, SUM(amount)
FROM sales
GROUP BY region
""")

for row in cursor.fetchall():
    print(row)
```

## 4. Statistical Analysis (Standard Deviation)

```python
cursor.execute("""
SELECT STDDEV(amount)
FROM sales
""")

result = cursor.fetchone()
print(f"Standard Deviation of Sales: {result[0]}")
```

## 5. Outlier Detection

```python
cursor.execute("""
SELECT *
FROM sales
WHERE amount > 
(SELECT AVG(amount) + 3*STDDEV(amount) FROM sales)
""")

outliers = cursor.fetchall()
if outliers:
    for row in outliers:
        print(row)
else:
    print("No outliers detected.")
```

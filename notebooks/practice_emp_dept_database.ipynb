## 1. Import Library
Loads MySQL Connector/Python to connect and interact with MySQL.

```python
import mysql.connector
```

## 2. Connect to MySQL Server
Connection objects used:
- conn: database connection
- cursor: executes SQL commands

```python
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="yourpassword"  # replace with your MySQL password
)

cursor = conn.cursor()
print("Connected to MySQL server!")
```

## 3. Create and Select Database

```python
cursor.execute("CREATE DATABASE IF NOT EXISTS employee_db")
cursor.execute("USE employee_db")
print("Database selected!")
```

## 4. Create DEPT Table

```python
cursor.execute("DROP TABLE IF EXISTS dept")
cursor.execute("""
CREATE TABLE dept (
    deptno INT PRIMARY KEY,
    dname VARCHAR(50),
    loc VARCHAR(50)
)
""")
print("Table 'dept' created!")
```

## 5. Create EMP Table

```python
cursor.execute("DROP TABLE IF EXISTS emp")
cursor.execute("""
CREATE TABLE emp (
    empno INT PRIMARY KEY,
    ename VARCHAR(50),
    job VARCHAR(50),
    mgr INT,
    hiredate DATE,
    sal DECIMAL(10,2),
    comm DECIMAL(10,2),
    deptno INT,
    FOREIGN KEY (deptno) REFERENCES dept(deptno)
)
""")
print("Table 'emp' created!")
```

## 6. Insert DEPT Rows

```python
dept_data = [
    (10, 'ACCOUNTING', 'NEW YORK'),
    (20, 'RESEARCH', 'DALLAS'),
    (30, 'SALES', 'CHICAGO'),
    (40, 'OPERATIONS', 'BOSTON')
]

cursor.executemany("INSERT INTO dept VALUES (%s, %s, %s)", dept_data)
conn.commit()
print("DEPT rows inserted!")
```

## 7. Insert EMP Rows

```python
emp_data = [
    (7369, 'SMITH', 'CLERK', 7902, '1980-12-17', 800, None, 20),
    (7499, 'ALLEN', 'SALESMAN', 7698, '1981-02-20', 1600, 300, 30),
    (7521, 'WARD', 'SALESMAN', 7698, '1981-02-22', 1250, 500, 30),
    (7566, 'JONES', 'MANAGER', 7839, '1981-04-02', 2975, None, 20),
    (7654, 'MARTIN', 'SALESMAN', 7698, '1981-09-28', 1250, 1400, 30),
    (7698, 'BLAKE', 'MANAGER', 7839, '1981-05-01', 2850, None, 30),
    (7782, 'CLARK', 'MANAGER', 7839, '1981-06-09', 2450, None, 10),
    (7788, 'SCOTT', 'ANALYST', 7566, '1987-07-13', 3000, None, 20),
    (7839, 'KING', 'PRESIDENT', None, '1981-11-17', 5000, None, 10),
    (7844, 'TURNER', 'SALESMAN', 7698, '1981-09-08', 1500, 0, 30),
    (7876, 'ADAMS', 'CLERK', 7788, '1987-07-13', 1100, None, 20),
    (7900, 'JAMES', 'CLERK', 7698, '1981-12-03', 950, None, 30),
    (7902, 'FORD', 'ANALYST', 7566, '1981-12-03', 3000, None, 20),
    (7934, 'MILLER', 'CLERK', 7782, '1982-01-23', 1300, None, 10)
]

query = """
INSERT INTO emp (empno, ename, job, mgr, hiredate, sal, comm, deptno)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
"""
cursor.executemany(query, emp_data)
conn.commit()
print("EMP rows inserted!")
```

## 8. View EMP Table

```python
cursor.execute("SELECT * FROM emp")
for row in cursor.fetchall():
    print(row)
```

## 9. Join EMP + DEPT

```python
cursor.execute("""
SELECT e.ename, e.job, d.dname, d.loc
FROM emp e
JOIN dept d ON e.deptno = d.deptno
""")
for row in cursor.fetchall():
    print(row)
```

## 10. Average Salary by Department

```python
cursor.execute("""
SELECT deptno, AVG(sal) AS avg_sal
FROM emp
GROUP BY deptno
""")
for row in cursor.fetchall():
    print(row)
```

## 11. Close Connection

```python
conn.close()
print("Connection closed!")
```
